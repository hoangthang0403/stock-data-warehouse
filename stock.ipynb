Fetching trade history data

```python
# Exploring VCB trade_history data
stock = Vnstock().stock(symbol='VCB', source='VCI')
stock.quote.history(start='2014-01-01', end='2024-09-30')
```

```python
# Defining a function to fetch all trade history
def fetching_trade_history(tickers):
    # Initializing stock object then fetch data into a DataFrame
    stock = Vnstock().stock(symbol='A32', source='VCI')
    stock = stock.quote.history(start='2024-09-01', end='2024-09-30')
    
    # Adding a ticker column
    stock.insert(0,'ticker','A32')
    
    # Empty a DataFrame
    trade_history = stock.drop(stock.index)
    for i in range(400):
        # Initializing stock object then fetch trade history data into a DataFrame
        stock = Vnstock().stock(symbol=tickers[i], source='VCI')
        stock = stock.quote.history(start='2014-01-01', end='2024-09-30')
        
        # Adding a ticker column
        stock.insert(0,'ticker',tickers[i])
        
        # Appending new trade history data
        trade_history = pd.concat([trade_history,stock])
        
    # Exporting data to CSV file
    # Modifying the file name after finished fetching 400 records
    trade_history.to_csv('trade_history400.csv',index=False) 
```

```python
# Fetching trade history data
fetching_trade_history(tickers)
```

```python
# Concatenating trade history
raw_trade_history400 = pd.read_csv('raw_trade_history400.csv')
raw_trade_history800 = pd.read_csv('raw_trade_history800.csv')
raw_trade_history1200 = pd.read_csv('raw_trade_history1200.csv')
raw_trade_history1598 = pd.read_csv('raw_trade_history1598.csv')
trade_history = pd.concat([raw_trade_history400,raw_trade_history800, raw_trade_history1200, raw_trade_history1598],ignore_index=True)
```

Fetching finance ratio data

```python
# Initializing stock object
stock = Vnstock().stock(symbol='VCI', source='VCI')

# Fetching finance ratio of VCI
finance_ratio = pd.DataFrame(stock.finance.ratio(period='year', lang='en'))
```

```python
# Exploring finance ratio data
finance_ratio
```

```python
# Exploring columns of DataFrame
finance_ratio.columns.values
```

```python
# Extracting inner_level index
cols = finance_ratio.columns.get_level_values(1)
cols = cols.values.tolist()
cols
```

```python
# Keeping necessary columns
cols = ['ticker',
    'yearReport',
    'Market Capital (Bn. VND)',
    'P/E',
    'P/B',
    'EPS (VND)',
    'BVPS (VND)',
    'ROE (%)',
    'ROA (%)',
    'Net Profit Margin (%)']
```

```python
def get_finance_ratio(tickers):
    # Initializing stock object then fetch data into a DataFrame
    stock = Vnstock().stock(symbol='VCI', source='VCI')
    stock = pd.DataFrame(stock.finance.ratio(period='year', lang='en'))
    
    # Extracting inner_level columns
    stock.columns = stock.columns.get_level_values(1)
    
    # Extracting necessary columns
    stock = stock[cols]
    finance_ratio = stock.drop(stock.index)
    
    for i in range(1000,1598):
        # Handling exception when there are no records for a specific company
        try:
            # Initializing stock object then fetch data into a DataFrame
            stock = Vnstock().stock(symbol=tickers[i], source='VCI')
            stock = pd.DataFrame(stock.finance.ratio(period='year', lang='en'))
            
            # Extracting inner_level columns then assign to DataFrame
            stock.columns = stock.columns.get_level_values(1)
        except KeyError:
            print(f'----------Missing {tickers[i]}"s finance_ratio---------')
            continue
        
        # Handling exception when there is a record doesn't have enough columns
        try:
            # Appending data to finance_ratio
            finance_ratio = pd.concat([finance_ratio,stock[cols]],ignore_index=True)
        except KeyError:
            print(f'----------{tickers[i]}---------')
    # Exporting data to CSV file
    # Modifying the file name after finished fetching 400 records
    finance_ratio.to_csv('finance_ratio1598.csv',index=False)
```

```python
get_finance_ratio(tickers)
```

```python
# Concatenating finance ratio data
raw_finance_ratio500 = pd.read_csv('raw_finance_ratio500.csv')
raw_finance_ratio1000 = pd.read_csv('raw_finance_ratio1000.csv')
raw_finance_ratio1598 = pd.read_csv('raw_finance_ratio1598.csv')
finance_ratio = pd.concat([raw_finance_ratio500,raw_finance_ratio1000,raw_finance_ratio1598],ignore_index=True)
```

```python
# Extracting finance ratio of companies since 2014
finance_ratio = finance_ratio[finance_ratio.yearReport >= 2014]

# Verifying result
finance_ratio.yearReport.min()
```

```python
# Exporting finance_ratio to CSV file
finance_ratio.to_csv('fact_finance_ratio.csv',index=False)
```

dim_date

```python
date_range = pd.date_range(start="2014-01-01", end="2024-09-30")
dim_date = pd.DataFrame({"date": date_range})
dim_date["year"] = dim_date["date"].dt.year
dim_date["month_id"] = dim_date["date"].dt.month
dim_date["quarter_id"] = dim_date["date"].dt.quarter
dim_date
```

```python
# Exporting dim_date to CSV file
dim_date.to_csv("dim_date.csv", index=False)
```

dim_quarter

```python
dim_quarter = dim_date[['quarter_id']].drop_duplicates().sort_values('quarter_id').reset_index(drop=True)
dim_quarter['quarter_name_short'] = dim_quarter['quarter_id'].apply(lambda x: f'Q{x}')
dim_quarter['quarter_name_long'] = dim_quarter['quarter_id'].apply(lambda x: f'Quarter {x}')
dim_quarter['quarter_number'] = dim_quarter['quarter_id']
dim_quarter
```

```python
# Exporting dim_quarter to CSV file
dim_quarter.to_csv("dim_quarter.csv", index=False)
```

dim_month

```python
dim_month = dim_date[['month_id']].drop_duplicates().sort_values('month_id').reset_index(drop=True)
dim_month['month_name_short'] = dim_month['month_id'].apply(lambda x: pd.to_datetime(f'2021-{x}-01').strftime('%b'))
dim_month['month_name_long'] = dim_month['month_id'].apply(lambda x: pd.to_datetime(f'2021-{x}-01').strftime('%B'))
dim_month['month_number'] = dim_month['month_id']
dim_month
```

```python
# Exporting dim_month to CSV file
dim_month.to_csv("dim_month.csv", index=False)
```
